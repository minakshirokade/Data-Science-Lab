```python
# AIM: Perform descriptive statistical analysis (mean, median, std, variance, skewness, kurtosis) to derive statistical insights from raw data.
```

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Perform descriptive statistical analysis
# Using .agg() to calculate the specific requested statistics
# Mean, median, standard deviation, variance, skewness, kurtosis
stats = df.agg(['mean', 'median', 'std', 'var', 'skew', 'kurt'])

print(stats)
```
