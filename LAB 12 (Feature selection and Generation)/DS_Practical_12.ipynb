```python
import pandas as pd
import numpy as np

# Sample dataset with numeric columns
data = {
    'age': [25, 30, 45, 35, 50],
    'income': [25000, 40000, 60000, 50000, 80000],
    'expenses': [20000, 25000, 40000, 35000, 60000]
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)

# -------------------------------
# 1️⃣ Mathematical / Statistical Features
# -------------------------------

# Mean, Median, Std Dev of numeric features (row-wise)
df['mean_value'] = df[['age', 'income', 'expenses']].mean(axis=1)
df['median_value'] = df[['age', 'income', 'expenses']].median(axis=1)
df['std_value'] = df[['age', 'income', 'expenses']].std(axis=1)

# Feature transformations
df['income_log'] = np.log1p(df['income'])      # log transform (handles skew)
df['income_sqrt'] = np.sqrt(df['income'])      # sqrt transform
df['income_square'] = df['income'] ** 2        # square transform

# Interaction features
df['income_expense_ratio'] = df['income'] / df['expenses']
df['income_minus_expense'] = df['income'] - df['expenses']

print("\nAfter Mathematical Feature Generation:")
print(df)
```

```python
import pandas as pd
from datetime import datetime, timedelta
import numpy as np

# ------------------------------
# Sample dataset with timestamp
# ------------------------------
data = {
    'user_id': [101, 102, 103, 104, 105],
    'purchase_date': [
        '2024-01-10 14:30:00',
        '2024-02-15 09:15:00',
        '2024-03-20 20:45:00',
        '2024-07-05 11:10:00',
        '2024-12-25 18:00:00'
    ],
    'amount': [2500, 3200, 1800, 4000, 5000]
}

df = pd.DataFrame(data)

# Convert purchase_date to datetime
df['purchase_date'] = pd.to_datetime(df['purchase_date'])

print("Original Dataset:")
print(df)

# ------------------------------
# Temporal Feature Generation
# ------------------------------

# 1️⃣ Extract year, month, day, hour
df['year'] = df['purchase_date'].dt.year
df['month'] = df['purchase_date'].dt.month
df['day'] = df['purchase_date'].dt.day
df['hour'] = df['purchase_date'].dt.hour

# 2️⃣ Day of week (0 = Monday, 6 = Sunday)
df['day_of_week'] = df['purchase_date'].dt.dayofweek

# 3️⃣ Weekend flag
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

# 4️⃣ Quarter of the year
df['quarter'] = df['purchase_date'].dt.quarter

# 5️⃣ Time of day bucket (morning, afternoon, evening, night)
def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['time_of_day'] = df['hour'].apply(time_of_day)

print("\nAfter Temporal Feature Generation:")
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# --------------------------------
# Sample dataset
# --------------------------------
data = {
    'Customer_ID': [1, 2, 3, 4, 5],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male'],
    'City': ['Mumbai', 'Delhi', 'Pune', 'Delhi', 'Chennai'],
    'Purchased': ['Yes', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)
print("Original Dataset:")
print(df)

# --------------------------------
# 1️⃣ Label Encoding
# --------------------------------
# Label Encoding converts each category into an integer
le = LabelEncoder()

df['Gender_Label'] = le.fit_transform(df['Gender'])
df['City_Label'] = le.fit_transform(df['City'])
df['Purchased_Label'] = le.fit_transform(df['Purchased'])

print("\nAfter Label Encoding:")
print(df[['Gender', 'Gender_Label', 'City', 'City_Label', 'Purchased', 'Purchased_Label']])

# --------------------------------
# 2️⃣ One Hot Encoding
# --------------------------------
# Each unique category becomes a separate binary column (0 or 1)
df_onehot = pd.get_dummies(df[['Gender', 'City']], drop_first=False)

print("\nAfter One Hot Encoding:")
print(df_onehot)

# Combine encoded features with original data
encoded_df = pd.concat([df, df_onehot], axis=1)

print("\nFinal Encoded DataFrame:")
print(encoded_df)
```

```python
# Temporal Feature Generation Example

import pandas as pd
import numpy as np

# Sample dataset with datetime column
data = {
    'timestamp': pd.date_range(start='2024-01-01', periods=10, freq='D')
}

df = pd.DataFrame(data)

# Generating temporal features
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
df['day_of_week'] = df['timestamp'].dt.dayofweek
df['hour'] = df['timestamp'].dt.hour
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
df['week_of_year'] = df['timestamp'].dt.isocalendar().week

print("✅ Temporal Feature Generation Complete!\n")
print(df)
```
